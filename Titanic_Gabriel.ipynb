```python
#dataframe
import pandas as pd
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
```

EJ 1

```python
print(df.shape)
print(df.dtypes)
```

EJ 2

```python
print("Edad media:", df["Age"].mean())
print("Tarifa media:",df["Fare"].mean())
```

EJ 3

```python
#sexo
print("Sexo de los pasajeros:", df["Sex"].value_counts())
#clase
print("Clase de los pasajeros:", df["Pclass"].value_counts())
#embarque
print("Embarcados:", df["Embarked"].value_counts())
```

EJ 4

```python
#valores faltantes
missing = df.isnull().sum()

#porcentaje de valores faltantes
missing_porcentaje = (missing / len(df)) * 100

print("Valores faltantes:", missing)
print("Porcentaje de valores faltantes:", missing_porcentaje)
```

EJ 5

```python
df['Deck'] = df['Cabin'].str[0]
print("Distribución de cubiertas por clase:")
print(pd.crosstab(df['Deck'], df['Pclass']))
```

EJ 6

```python
#completamente duplicadas
duplicados_completos = df.duplicated().sum()
#mismo nombre y ticket
duplicados_parciales = df.duplicated(subset=['Name','Ticket']).sum()
print("Duplicados completos:", duplicados_completos)
print("Duplicados parciales:", duplicados_parciales)
```

EJ 7

```python
#verificar Pclass
print("Pclass:", df['Pclass'].value_counts())
#edad negativa
print("Edad negativa:", df[df['Age'] < 0])
#tarifa negativa
print("Tarifa negativa:", df[df['Fare'] < 0])
```

EJ 8

```python
#Cogemos valores antes de un punto "."
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)
medianas_por_titulo = df.groupby('Title')['Age'].median()
print("Medianas de edad por título:")
print(medianas_por_titulo, df.groupby('Title')['Age'].count())

df['Age'] = df['Age'].fillna(df.groupby('Title')['Age'].transform('median'))
if df['Age'].isnull().any():
    df['Age'] = df['Age'].fillna(df['Age'].median())
print(f"Valores faltantes en Age: {df['Age'].isnull().sum()}")
```

EJ 9

```python
porcentaje_de_faltantes = (df['Cabin'].isnull().sum() / len(df))
print(f"Porcentaje de valores que faltan en Cabin: {porcentaje_de_faltantes:.2%}")

missing_cabin = df['Cabin'].isna().sum()
porcentaje_missing = missing_cabin / len(df) * 100

print(f"Valores que faltan en Cabin: {missing_cabin}")
print(f"Porcentaje de missing en Cabin: {porcentaje_missing:.2f}%")

df['HasCabin'] = df['Cabin'].notna().astype(int)
print(df['HasCabin'].value_counts())
```

EJ 10

```python
print(f"Valores que faltan en Embarked: {df['Embarked'].isna().sum()}")

moda_embarked = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(moda_embarked)
print(df['Embarked'].value_counts())
```

EJ 11

```python
supervivencia_por_sexo = df.groupby('Sex')['Survived'].mean() * 100
print(supervivencia_por_sexo)
```

EJ 12

```python
supervivencia_clase = df.groupby('Pclass')['Survived'].mean() * 100
print(supervivencia_clase)
```

EJ 13

```python
tabla_sexo_clase = pd.crosstab(
    df['Sex'],
    df['Pclass'],
    values=df['Survived'],
    aggfunc='mean'
) * 100

print(tabla_sexo_clase)
```

EJ 14

```python
import pandas as pd
# Crear grupos de edad
bins = [0, 12, 17, 59, df['Age'].max()]
labels = ['Niño', 'Adolescente', 'Adulto', 'Mayor']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

# Calcular tasa de supervivencia por grupo
supervivientes_por_edades = df.groupby('AgeGroup', observed=False)['Survived'].mean()
print(supervivientes_por_edades)
```

EJ 15

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
def clasificar_familia(size):
    if size == 1:
        return 'Solo'
    elif 2 <= size <= 4:
        return 'Pequeña'
    else:
        return 'Grande'
df['TipoFamilia'] = df['FamilySize'].apply(clasificar_familia)
supervivencia_familia = df.groupby('TipoFamilia')['Survived'].mean() * 100
print(supervivencia_familia)
```
